```python
from queue import PriorityQueue

def best_first_search(graph, start_state, end_state):
    visited = set()
    queue = PriorityQueue()
    queue.put((0, start_state))
    parent = {}
    path = []
    steps = []


    while not queue.empty():
        _, current_state = queue.get()
        if current_state == end_state:
            path.append(current_state)
            while current_state in parent:
                current_state = parent[current_state]
                path.append(current_state)
            path.reverse()
            steps.append([end_state, 'ket thuc', ''])
            break

        visited.add(current_state)
        steps.append([current_state, graph[current_state], list(visited)])

        for neighbor in graph[current_state]:
            if neighbor not in visited:
                queue.put((heuristic(neighbor, end_state), neighbor))
                parent[neighbor] = current_state

    return steps, path

def heuristic(state, end_state):
    return int(state.split('(')[1].split(')')[0])

# Read the graph, start state, and end state from the input
input_file_path = 'input.txt'

with open(input_file_path, 'r', encoding='utf-8') as file:
    code = file.read()

variables = {}
exec(code, variables)

graph = variables.get('graph')
start_state = variables.get('start_state')
end_state = variables.get('end_state')

# Apply the Best-First Search algorithm
steps, path = best_first_search(graph, start_state, end_state)

output_file_path = 'output.txt'

with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write("Bảng liệt kê các bước thực hiện thuật toán:\n")
    file.write("+" + "-" * 37 + "+" + "-" * 37 + "+" + "-" * 37 + "+\n")
    file.write("| {:^35s} | {:^35s} | {:^35s} |\n".format("Phát triển thuật toán", "Trạng thái kề", "Danh sách L"))
    file.write("+" + "-" * 37 + "+" + "-" * 37 + "+" + "-" * 37 + "+\n")
    for step in steps:
        file.write("| {:<35s} | {:<35s} | {:<35s} |\n".format(step[0], str(step[1]), str(step[2])))
    file.write("+" + "-" * 37 + "+" + "-" * 37 + "+" + "-" * 37 + "+\n\n")
    file.write("Đường đi từ Trạng thái đầu đến Trạng thái kết thúc:\n")
    file.write(" => ".join(path))
```
